```python
import re
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Food_Inspections.csv")
```

```python
df.shape
```

```python
df = df.loc[df['Results'] == 'Fail']
```

```python
df.shape
```

```python
df.head(10)
```

```python
df.dropna(subset=["Violations"], inplace=True)
```

```python
df.shape
```

```python
df.head(10)
```

```python
df_violations = df['Violations']
df_violations.head(10)
```

```python
dic = {}
for i in df_violations.values:
    if not pd.isnull(i): #### remove the NaNs in this pandas series of violations data
        desc = i.split(' | ') #### Reasons for the violations are separated by |
        for txt in desc: ### for each Reason
            if re.match(r'^[0-9]+\.', txt): # the reason always starts with numerical digits with a dot
                reason = txt.split(' - Comments')[0] # removing the comments section in the reasons
                if reason in dic: dic[reason] += 1
                else: dic[reason] = 1
```

```python
dic
```

```python
## sorting the violation reasons with respect to the counts
for key, value in sorted(dic.items(), key=lambda k_v: k_v[1], reverse=True)[:10]:
    print ("%s: %s" % (key, value))
```
